# Linear Regression

Independent variable - Which affect the dependent variable(many), factors determining ddependent variable.

Dependent variable - The variable whose value depend on independent variables.(one)

Variables can be Continuous or Categorical.

But we need to convert Categorical into Numerical.

Here we want to determine height from the given weight.

height = a + b * weight

```python
weight = [62,65,66,71,74,62,66,61,56,72,75,60]
height = [156,160,163,170,175,155,162,153,145,170,175,150]
```

```python
len(weight), len(height)
```

####  Convert data into a dataframe 

Because machine learning can only be performed on a Dataframe

```python
import pandas

weight_df = pandas.DataFrame(weight)
height_df = pandas.DataFrame(height)
```

```python
height_df
```

```python
weight_df
```

### Create a Model

Model hepls you get the best value of intercept and coefficients(during the training process) 

```python
from sklearn import linear_model # Avoid importing whole library 
                                # beacuse it is very big in size
                                # so import only a function from this library
```

```python
model = linear_model.LinearRegression()        # empty model creation
```

```python
model.fit(weight_df,height_df)  # on fitting weight and height into the model
                              # it finds the relationship between height and weight
                              # i.e the value of a and b
```

```python
model.coef_          # b(coeffient) in------------- y = a + bx 
```

```python
model.intercept_     # a(intercept) in------------- y = a + bx
```

### Predicting height for a new person

```python
# for given weight 60 we want to find the height of a person

54.37 + 1.62 * 60
```

### Predicting height for many new people

```python
new_people = [54,60,73]
```

```python
new_people_df = pandas.DataFrame(new_people)
```

```python
model.predict(new_people_df)
```

### Find accuracy of prediction(model)

```python
model.score(weight_df,height_df)
```

### Precision and Recall

**Precision of one** = correctly classified ones / total actual ones


**Recall of one** = correctly classified ones / total predicted ones

```python
69/71
```

**Confusion matrix** is a n x n matrix which is used to evaluate the performance of a classification model.

The order of the confusion matrix = (the no. of categories in the dependent variable)^2

The confusion matrix for a binary problem can be written as 


```python
#                           Actual

#                               Positive           Negative
#                            ____________________________________   
#                           |
# Predicted     Positive    |    TP                   FP
#                           |
#               Negative    |    FN                   TN
#                           | 
#                           |

# (outcome = TRUE or False)(Predicted = Positive or Negative)
```

**F1 Score** is the harmonic mean between Precision and Recall.

```python
f1_score = 2 / ((1/0.84) + (1/ 0.95))    # 2a*b / a+b
f1_score
```

There are 3 ways to determine the optimum no. of factors/dimensions.

     1) eigen_values > 1:

     There should be 3 dimensions(because 3 values are satisfying above condition).

     2) Scree Plot:
     Choose elbow point as no. of factors, here 2, 3 or 4.

     3) Cummulative variance >= 60%

Variance = % explained/contribution by each factor.

Max. Variance = 1 

On reducing factors, Variance decreases. So we need to make a balance between variance and min no. of factors.


     The factor variance function returns 3 arrays
    1st = Variance - contribution of each factor
    2nd = proportional variance - % contribution of each factor.
    3rd = Cummulative variance  - cummulative sum of proportional variance.   (sum = 1)

    Find min no. of factors giving cummulative variance > 0.6

Since 3 and 4 factors both give >60% cummulative variance so we will choose min 
i.e

 factor = 3 is optimum.
